```python
min = float(df["Worldwide Gross"][0].replace('$', ''))
max = float(df["Worldwide Gross"][0].replace('$', ''))
minindex = 0
maxindex = 0
for i in range(0, 77):
    if (min > float(df["Worldwide Gross"][i].replace('$', ''))):
        minindex = df["Worldwide Gross"][i]
        min = float(df["Worldwide Gross"][i].replace('$', ''))
    if (max < float(df["Worldwide Gross"][i].replace('$', ''))):
        maxindex = df["Worldwide Gross"][i]
        max = float(df["Worldwide Gross"][i].replace('$', ''))

df[(df['Worldwide Gross'] == maxindex) | (df['Worldwide Gross'] == minindex)]
```

```python
tmplist = []
for i in range(0, 77):
    tmplist.append(str(df['Film'][i][0]) + str(df['Year'][i]))
df.assign(FilmCode=tmplist)
```

```python
if (type(df["Worldwide Gross"][0]) == str):
    for i in range (0, 77):
        df["Worldwide Gross"][i] = float(df["Worldwide Gross"][i].replace('$', ''))
df["Worldwide Gross"] = df["Worldwide Gross"].astype('float')
df.corr()
```
