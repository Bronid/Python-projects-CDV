```python
from sklearn import datasets
import pandas as pd

dataset = datasets.load_iris()

df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df["Target"] = dataset['target']
df = df[['petal width (cm)', 'Target']]
df = df[df['Target'] < 2]
df
```

```python
import math as m
lr = 0.01

# Log loss function
def LogicznaRegresja(X, y, lr, n):
    a = 0
    b = 0
    p = len(X)
    for l in range(n):
        mse = 0
        pochodneA = 0
        pochodneB = 0
        for i in range(p):
            pred = 1 / (1 + m.exp(-(a * X[i] + b)))
            mse -= y[i] * m.log(pred) + (1 - y[i]) * m.log(1 - pred)
            pochodneA += X[i] * ((pred) - y[i])
            pochodneB += (pred) - y[i]
        mse = mse / p
        pochodneA = pochodneA / p
        pochodneB = pochodneB / p
        a = a - (lr * pochodneA)
        b = b - (lr * pochodneB)
        print(mse)
    print(mse)
    print(a)
    print(b)
       
LogicznaRegresja(df["petal width (cm)"], df["Target"], lr, 10000)
```
