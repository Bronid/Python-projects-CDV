```python
## Zad. 18 Proszę sprawdzić, czy w w df znajdują się osoby, które mają taki sam numer biletu. Jeśli tak - proszę podzielić Fare przez odpowiednią liczbę osób (opłata za rejs) i zapisać w kolumnie FareCorr
repeatcount = 0
list = []
for a in range(0, df.shape[0]):
    for i in range(0, df.shape[0]):
        if (df['Ticket'][a] == df['Ticket'][i]): 
            repeatcount += 1
    list.append(df['Fare'][a] / repeatcount)
    repeatcount = 0

df['FareCorr'] = list

df
```

```python
## Zad. 19 Proszę utworzyć nową kolumnę FareCat z etykietą per Fare "low" ,"medium", "high" w zależności od opłaty biletu 0,15, 65 
list = []

for fare in df['Fare']:
    if (fare < 15): list.append("low")
    elif (fare >= 15 and fare < 65): list.append("medium")
    else: list.append("high")

df['FareCat'] = list
df
```

```python
## Zad. 22 Proszę utworzyć nową kolumnę "tytuł", gdzie znajdą się np. Mr, Miss, Dr, Master
df['Tytul'] = df['Name'].str.extract('([A-Za-z]+\.)')

df
```

```python
## Zad. 23 Na podstawie tytułu proszę uzupełnić wiersze, które mają puste pola w Age oraz AgeCat. Dla każdego z tytułu proszę policzyć medianę i taką wartością uzupełnić Nulle
medianAge = df.groupby('Tytul', group_keys=True).Age.median()
df.Age = df.groupby('Tytul', group_keys=True).Age.apply(lambda x: x.fillna(x.median()))
df['AgeCat'] = pd.cut(df['Age'], bins=[0,18,100], labels=['child','adult'])
```

```python
list = []
listagecat = []
for a in range(0, df.shape[0]):
    if (df['Age'].isnull()[a]):
        list.append(medianAge[df["Tytul"][a]])
    else: list.append(df['Age'][a])
df['Age'] = list

list = []
for age in df['Age']:
    if (age >= 18): list.append("adult")
    elif (age < 18): list.append("child")
    else: list.append(None)
df['AgeCat'] = list

df
```

```python
## Zad. 24 Z kolumny tytuł proszę sprawdzić jaka głównie występuje płeć i uzupełnić Nulle w kolumnie płeć
```

```python
## Zad. 25 Proszę utworzyć kolumnę familySize, gdzie nadadzą Państwo klasę "small", "medium", "big" na podstawie Parch, oraz Sib / Sp
df['FamilySize'] = df.apply(lambda row: 'small' if (row['Parch'] + row['SibSp'] == 0) 
                            else ('medium' if (row['Parch'] + row['SibSp'] > 0) and (row['Parch'] + row['SibSp'] <= 4)
                                 else 'big'), axis=1)

df
```

```python
## Zad. 26 Na podstawie utworzonej kolumny w zadaniu 25 proszę o utworzenie wykresu, którzy przedstawi szanse na przeżycie
sns.countplot(x='FamilySize', hue='Survived', data=df)
```

```python
## Zad. 27 Proszę uzupełnić kolumnę Nulle w Embarked najczęstszą wartością występującą w df
most_frequent = df['Embarked'].value_counts().idxmax()
df['Embarked'].fillna(most_frequent, inplace=True)

df
```

```python
df['Age'] = pd.qcut(df['Age'], q = 7)
df
```

```python
df['FareCorr'] = pd.qcut(df['FareCorr'], q = 7)
df
```

```python
from sklearn.preprocessing import OneHotEncoder

df['Sex'] = df['Sex'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Age'] = df['Age'].astype('category')
df['Sex'] = df['Sex'].cat.codes
df['Pclass'] = df['Pclass'].cat.codes
df['Age'] = df['Age'].cat.codes

df
```

```python
df.isna().sum()

del df['Cabin']
del df['Name']
del df['PassengerId']
del df['Ticket']

df.isna().sum()
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked', 
'AgeCat', 'FareCorr', 'FareCat', 'Tytul', 'FamilySize'])
one_hot_encoded_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
one_hot_encoded_data['Fare'] = scaler.fit_transform(one_hot_encoded_data[['Fare']])
one_hot_encoded_data
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers

model = keras.Sequential()
model.add(layers.Dense(32, input_dim=62, activation="sigmoid", name="layer1"))
model.add(layers.Dense(64, activation="sigmoid", name="layer2"))
model.add(layers.Dense(1, activation="sigmoid", name="layer3"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics = ['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X = one_hot_encoded_data.drop(['Survived'], axis=1)
y = one_hot_encoded_data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
model.fit(X_train, y_train, epochs = 800)
model.predict(X_test)
```
