```python
## Zad. 18 ProszÄ™ sprawdziÄ‡, czy w w df znajdujÄ… siÄ™ osoby, ktÃ³re majÄ… taki sam numer biletu. JeÅ›li tak - proszÄ™ podzieliÄ‡ Fare przez odpowiedniÄ… liczbÄ™ osÃ³b (opÅ‚ata za rejs) i zapisaÄ‡ w kolumnie FareCorr
repeatcount = 0
list = []
for a in range(0, df.shape[0]):
    for i in range(0, df.shape[0]):
        if (df['Ticket'][a] == df['Ticket'][i]): 
            repeatcount += 1
    list.append(df['Fare'][a] / repeatcount)
    repeatcount = 0

df['FareCorr'] = list

df
```

```python
## Zad. 19 ProszÄ™ utworzyÄ‡ nowÄ… kolumnÄ™ FareCat z etykietÄ… per Fare "low" ,"medium", "high" w zaleÅ¼noÅ›ci od opÅ‚aty biletu 0,15, 65 
list = []

for fare in df['Fare']:
    if (fare < 15): list.append("low")
    elif (fare >= 15 and fare < 65): list.append("medium")
    else: list.append("high")

df['FareCat'] = list
df
```

```python
## Zad. 20 Za pomocÄ… sns wyÅ›wietl wykres sÅ‚upkowy, ktÃ³ry przestawi prawdopodobieÅ„stwo przeÅ¼ycia per FareCat + sÅ‚owne wnioski
sns.barplot(x=df['FareCat'], y=df['Survived'])
```

```python
## Zad. 21 Za pomocÄ… sns wyÅ›wietl wykres sÅ‚upkowy, ktÃ³ry przestawi prawdopodobieÅ„stwo przeÅ¼ycia per Embarked (port z ktÃ³rego wsiadaÅ‚ pasaÅ¼er)
sns.barplot(x=df['Embarked'], y=df['Survived'])
```

```python
## Zad. 22 ProszÄ™ utworzyÄ‡ nowÄ… kolumnÄ™ "tytuÅ‚", gdzie znajdÄ… siÄ™ np. Mr, Miss, Dr, Master
df['Tytul'] = df['Name'].str.extract('([A-Za-z]+\.)')

df
```

```python
## Zad. 23 Na podstawie tytuÅ‚u proszÄ™ uzupeÅ‚niÄ‡ wiersze, ktÃ³re majÄ… puste pola w Age oraz AgeCat. Dla kaÅ¼dego z tytuÅ‚u proszÄ™ policzyÄ‡ medianÄ™ i takÄ… wartoÅ›ciÄ… uzupeÅ‚niÄ‡ Nulle
medianAge = df.groupby('Tytul', group_keys=True).Age.median()
df.Age = df.groupby('Tytul', group_keys=True).Age.apply(lambda x: x.fillna(x.median()))
df['AgeCat'] = pd.cut(df['Age'], bins=[0,18,100], labels=['child','adult'])
```

```python
list = []
listagecat = []
for a in range(0, df.shape[0]):
    if (df['Age'].isnull()[a]):
        list.append(medianAge[df["Tytul"][a]])
    else: list.append(df['Age'][a])
df['Age'] = list

list = []
for age in df['Age']:
    if (age >= 18): list.append("adult")
    elif (age < 18): list.append("child")
    else: list.append(None)
df['AgeCat'] = list

df
```

```python
## Zad. 24 Z kolumny tytuÅ‚ proszÄ™ sprawdziÄ‡ jaka gÅ‚Ã³wnie wystÄ™puje pÅ‚eÄ‡ i uzupeÅ‚niÄ‡ Nulle w kolumnie pÅ‚eÄ‡
```

```python
## Zad. 25 ProszÄ™ utworzyÄ‡ kolumnÄ™ familySize, gdzie nadadzÄ… PaÅ„stwo klasÄ™ "small", "medium", "big" na podstawie Parch, oraz Sib / Sp
df['FamilySize'] = df.apply(lambda row: 'small' if (row['Parch'] + row['SibSp'] == 0) 
                            else ('medium' if (row['Parch'] + row['SibSp'] > 0) and (row['Parch'] + row['SibSp'] <= 4)
                                 else 'big'), axis=1)

df
```

```python
## Zad. 26 Na podstawie utworzonej kolumny w zadaniu 25 proszÄ™ o utworzenie wykresu, ktÃ³rzy przedstawi szanse na przeÅ¼ycie
sns.countplot(x='FamilySize', hue='Survived', data=df)
```

```python
## Zad. 27 ProszÄ™ uzupeÅ‚niÄ‡ kolumnÄ™ Nulle w Embarked najczÄ™stszÄ… wartoÅ›ciÄ… wystÄ™pujÄ…cÄ… w df
most_frequent = df['Embarked'].value_counts().idxmax()
df['Embarked'].fillna(most_frequent, inplace=True)

df
```

```python
df['Age'] = pd.qcut(df['Age'], q = 7)
df
```

```python
df['FareCorr'] = pd.qcut(df['FareCorr'], q = 7)
df
```

```python
from sklearn.preprocessing import OneHotEncoder

df['Sex'] = df['Sex'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Age'] = df['Age'].astype('category')
df['Sex'] = df['Sex'].cat.codes
df['Pclass'] = df['Pclass'].cat.codes
df['Age'] = df['Age'].cat.codes

df
```

```python
df.isna().sum()

del df['Cabin']
del df['Name']
del df['PassengerId']
del df['Ticket']

df.isna().sum()
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked', 
'AgeCat', 'FareCorr', 'FareCat', 'Tytul', 'FamilySize'])
one_hot_encoded_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
one_hot_encoded_data['Fare'] = scaler.fit_transform(one_hot_encoded_data[['Fare']])
one_hot_encoded_data
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers

model = keras.Sequential()
model.add(layers.Dense(32, input_dim=62, activation="sigmoid", name="layer1"))
model.add(layers.Dense(64, activation="sigmoid", name="layer2"))
model.add(layers.Dense(1, activation="sigmoid", name="layer3"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics = ['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X = one_hot_encoded_data.drop(['Survived'], axis=1)
y = one_hot_encoded_data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
model.fit(X_train, y_train, epochs = 800)
model.predict(X_test)
```
